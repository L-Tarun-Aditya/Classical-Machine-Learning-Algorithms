```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
class ltaLR:
    def __init__(self):
        self.m = None
        self.b = None
    def fit(self, x_train, y_train):
        x = np.array(x_train)
        y = np.array(y_train)


        mean_x = np.mean(x)
        mean_y = np.mean(y)

        numerator = sum((x- mean_x)*(y- mean_y))
        denominator = sum((x - mean_x)**2)

        self.m =  numerator/denominator
        self.b = mean_y - self.m*mean_x
        
        
    def predict(self, x_test):
        x = np.array(x_test)

        predictions = self.m * x + self.b
        return predictions
               
```

```python
df = pd.read_csv("placement.csv")
df.head()

x = df.iloc[:,0]
y = df.iloc[:, 1]
```

```python
x_train, x_test, y_train, y_test= sk.sklearn.model_selection.train_test_split(x,y, test_size=0.2, train_size=0.8, random_state=2, shuffle=True, stratify=None)
```

```python
obj = ltaLR()
obj.fit(x_train, y_train)

```

```python
tempvar = obj.predict(x_test)
print(tempvar)
print(y_test)
```
